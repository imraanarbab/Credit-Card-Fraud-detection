```python
import pandas as pd
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
Q1 = df.Amount.quantile(0.25)
Q3 = df.Amount.quantile(0.75)
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
#Upper and lower limits
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
#Showing all the outliers
df[(df.Amount<lower_limit)|(df.Amount>upper_limit)]
```

```python
df_no_outlier = df[(df.Amount>lower_limit)&(df.Amount<upper_limit)]
df_no_outlier
```

```python
df_no_outlier.shape
```

```python
Q1 = df_no_outlier.Time.quantile(0.25)
Q3 = df_no_outlier.Time.quantile(0.75)
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
#Upper and lower limits
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
df_no_outlier2 = df_no_outlier[(df_no_outlier.Time>lower_limit)&(df_no_outlier.Time<upper_limit)]
df_no_outlier2
```

```python
df_no_outlier2.describe()
```

```python
# from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC  # Import the SVC class from sklearn.svm
```

```python
model_params = {
    'svm': {
        'model': SVC(),
        'params': {
            'C': [1, 10, 20],
            'kernel': ['rbf', 'linear', 'poly', 'sigmoid']
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [1, 5, 80],
            'criterion':['entropy', 'gini']
        }
    },
    'logistic_regression': {
        'model': LogisticRegression(solver='liblinear', multi_class='auto'),
        'params': {
            'C': [1, 5, 20]
        }
    },
    'gaussian_nb': {
        'model': GaussianNB(),
        'params': {}
    },
    'decision_tree_classifier': {
        'model': DecisionTreeClassifier(),
        'params': {
            'criterion': ['entropy', 'gini']
        }
    }
}
```

```python
X = df_no_outlier2.drop(['Class'], axis = 'columns')
```

```python
y = df_no_outlier2.Class
```

```python
scores = []
for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(X, y)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })

# Create a DataFrame with the results
df = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
print(df)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
X_train
```

```python
from sklearn.svm import SVC
model = SVC(C = 1, kernel = 'rbf')
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
